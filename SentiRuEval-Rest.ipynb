```python
import csv
from collections import Counter
reader = csv.reader(open("eval_В_rest.csv", "r"), delimiter="\t")
next(reader)
next(reader)
next(reader)
errors = Counter()
preds = Counter()
for row in reader:
    actual, pred = row[1], row[2]
    preds[pred] += 1 
    if actual != pred:
        errors[actual+"-"+pred] += 1
print(preds)
print(errors)
```
